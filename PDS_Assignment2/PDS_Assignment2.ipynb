```python
import pandas as pd
```

```python
df=pd.read_csv("train.csv")
```

```python
df.head()
```

```python
#Checking for null values
```

```python
df.isna().sum()
```

```python
df2 = pd.DataFrame(df)
```

```python
# (b) Remove the units from some of the attributes and only keep the numerical values (for example remove kmpl from “Mileage”, CC from “Engine”, bhp from “Power”, and lakh from “New_price”).
```

```python
df2['Mileage'] = df2['Mileage'].str.extract('(\d+\.\d+)').astype(float)

df2['Engine'] = df2['Engine'].str.replace(' CC', '').astype(float)

df2['Power'] = df2['Power'].str.extract('(\d+\.\d+)').astype(float)

df2['New_Price'] = df2['New_Price'].str.extract('(\d+\.\d+)').astype(float)
```

```python
# (a)Look for the missing values in all the columns and either impute them (replace with mean,median, or mode) or drop them. Justify your action for this task
```

```python
df2['Mileage'].fillna(df2['Mileage'].median(),inplace=True)
df2['Engine'].fillna(df2['Engine'].median(),inplace=True)
df2['Power'].fillna(df2['Power'].median(),inplace=True)
df2['Seats'].fillna(df2['Seats'].mode()[0],inplace=True)
df2.drop('New_Price', axis=1, inplace=True)#axis=1 indicates to drop a column


print(df2.head())
```

```python
opt_csv = 'clean_data.csv'
df2.to_csv(opt_csv, index=False)
```
